```python
# Trasformazione e Predizione
# 1.
skdataset = dataset.copy(deep=True)

# Rimuovo il campo 'Date' in quanto stringa e non rilevante ai fini della predizione
skdataset = skdataset.drop(columns=['Date'])

# Aggiorno il campo 'Mean_Temperature_F' in quanto presenta un valore nullo
dataset.loc[dataset['Mean_Temperature_F'].isnull(), 'Mean_Temperature_F'] = (skdataset['Max_Temperature_F'] - skdataset['Min_TemperatureF']) / 2

# Sostituisco i valori nel campo 'Events' con 1 se si è verificato un evento, con 0 altrimenti
skdataset.loc[skdataset['Events'].notnull(), 'Events'] = 1
skdataset.loc[skdataset['Events'].isnull(), 'Events'] = 0
```

```python
# 2. Divido il dataset skdataset in 'test' e 'train' composti dai 2/3 e 1/3 delle righe rispettivamente
row_boundary = int(2/3 * len(skdataset.index))
train = skdataset.loc[:row_boundary, :]
test = skdataset.loc[row_boundary:, :]
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

preprocessing_transformer = Pipeline(steps=[('std_scale', StandardScaler())])
model_bnb = Pipeline(steps=[('bernoulli_nb', BernoulliNB())])
bernoulli_pipeline = Pipeline(steps=[('preprocessing', preprocessing_transformer),
                                    ('model', model_bnb)])
```

```python
train_data = train.loc[:, train.columns != 'Events']
train_target = train.loc[:, train.columns == 'Events']
test_data = test.loc[:, test.columns != 'Events']
test_target = test.loc[:, test.columns == 'Events']
```

```python
trX_train, trX_test, try_train, try_test = train_test_split(train_data, train_target, train_size=0.8, test_size=0.2)
teX_train, teX_test, tey_train, tey_test = train_test_split(test_data, test_target, train_size=0.8, test_size=0.2)
```

```python
bernoulli_pipeline.fit(trX_train, try_train.values.ravel())
train_pred = bernoulli_pipeline.predict(trX_test)
bernoulli_pipeline.fit(teX_train, tey_train.values.ravel())
test_pred = bernoulli_pipeline.predict(teX_test)

train_score = accuracy_score(try_test, train_pred)
test_score = accuracy_score(tey_test, test_pred)
```

```python
# 3. Valori di accuratezza
print(f'BernoulliNB su train dataset: accuracy {train_score}')
print(f'BernoulliNB su test dataset: accuracy {test_score}')
```

```python
# 4. Analisi delle confusion matrix
print(f'Confusion matrix su train dataset')
print(confusion_matrix(try_test, train_pred))
print(f'Confusion matrix su test dataset')
print(confusion_matrix(tey_test, test_pred))
```

Le predizioni su entrambi i dataset sono ben bilanciate sia nelle predizioni della classe 0 (nessun evento) sia nelle predizioni della classe 1 (evento verificato)

```python
# 5. Valutare le prediction tramite DecisionTree

from sklearn.tree import DecisionTreeClassifier
model_dt = Pipeline(steps=[('decision_tree', DecisionTreeClassifier())])
dt_pipeline = Pipeline(steps=[('model', model_dt)])

dt_pipeline.fit(trX_train, try_train.values.ravel())
dt_train_pred = dt_pipeline.predict(trX_test)
dt_pipeline.fit(teX_train, tey_train.values.ravel())
dt_test_pred = dt_pipeline.predict(teX_test)

dt_train_score = accuracy_score(try_test, dt_train_pred)
dt_test_score = accuracy_score(tey_test, dt_test_pred)
```

```python
print(f'DecisionTree su train dataset: accuracy {dt_train_score}')
print(f'DecisionTree su test dataset: accuracy {dt_test_score}')
```

```python
print(f'Confusion matrix su train dataset con DecisionTree')
print(confusion_matrix(try_test, dt_train_pred))
print(f'Confusion matrix su test dataset con DecisionTree')
print(confusion_matrix(tey_test, dt_test_pred))
```

Con l'utilizzo di un DecisionTree la precisione risulta inferiore in entrambi i dataset di circa il 6%. Per quanto riguarda le confusion matrix, le predizioni sul dataset train presentano un maggior numero di falsi positivi mentre le predizioni sul dataset test presentano un maggior numero di falsi negativi. In entrambi i casi (coerentemente con la diminuzione della precisione) si registra una sensibile riduzione di true positive e true negative.
